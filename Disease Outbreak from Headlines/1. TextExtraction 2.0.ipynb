## Create geography list

```python
gc = geonamescache.GeonamesCache()

countries = gc.get_countries_by_names()
country_list = [unidecode.unidecode(x) for x in countries.keys()] # country names list
country_list = sorted(list(set(country_list)), key=lambda x: len(x), reverse=True)

cities = gc.get_cities()
city_list = []
for k,v in cities.items():
    city_list.append(unidecode.unidecode(v['name'])) # city names list
city_list = sorted(list(set(city_list)), key=lambda x: len(x), reverse=True)

cities2 = dict() # City name as key
for k in list(cities.keys()):
    cities2[unidecode.unidecode(cities[k]['name']).lower()] = cities[k]
    # turn them into lower cases to avoid extraction problem later due to case sensitivity.

countries2 = dict() # Country code as key
for k in list(countries.keys()):
    countries2[countries[k]['iso']] = countries[k]
```

## Furthur Macth to counties and States: fill out the missing info

```python
# Get the na list
co = list(hl2.index[hl2.country.isna()])
ci = list(hl2.index[hl2.city.isna()])

states = gc.get_us_states()
state_list = sorted([unidecode.unidecode(states[x]['name']) for x in states], key = len, reverse = True)

states2 = dict() # Use state name as key
for k in list(states.keys()):
    states2[states[k]['name']] = states[k]

counties = gc.get_us_counties()
counties2 = dict() # Use county name as key
for k in counties:
    counties2[k['name']] = k
county_list = sorted([unidecode.unidecode(x) for x in list(counties2.keys())], key = len, reverse = True)

compile_state = re.compile(r'\b' + r'\b|\b'.join(state_list) + r'\b')#, flags=re.IGNORECASE)
compile_county = re.compile(r'\b' + r'\b|\b'.join(county_list) + r'\b')#, flags=re.IGNORECASE)
drop = 0
drop_index = []
for ix in range(hl2.shape[0]):
    try:
        # print(ix)
        h = hl2.headline[ix]
        if ix not in co: # has country name, don't have to match further
            continue
        elif ix not in ci: # has city but not country name: find country
            hl2['country'][ix] = countries2[cities2[hl2.city[ix].lower()]['countrycode']]['name']
        else:               # has no city or country name: find state or county (only for the US)
            res = re.search(compile_state, h).group(0) if re.search(compile_state, h) else None # find state
            if not res:     # no state -> find county
                res = re.search(compile_county, h).group(0) if re.search(compile_county, h) else None
            
            if not res:     # no state or county
                print(ix, hl2.headline[ix])
                drop += 1
                drop_index.append(ix)
            else:
                hl2.city[ix] = res
                hl2.country[ix] = 'United States'
    except:
        print('error', ix, hl2.headline[ix])
        drop += 1
print('There\'s',drop,'rows which can\'t be recognized in geospacecache.')
```

```python
# cities
def gen_dict_extract(var, key):
    if isinstance(var, dict):
        for k, v in var.items():
            if k == key:
                yield v
            if isinstance(v, (dict, list)):
                yield from gen_dict_extract(v, key)
    elif isinstance(var, list):
        for d in var:
            yield from gen_dict_extract(d, key)
cities = [*gen_dict_extract(cities, 'name')]
len(cities)
```

```python
w = "TEMPLATES = ( ('index.html', 'home'), ('base.html', 'base'))"

# find outer parens
outer = re.compile("\((.+)\)")
mm = outer.search(w) # match="( ('index.html', 'home'), ('base.html', 'base'))" will match the most outter one
inner_str = mm.group(1) # ('index.html', 'home'), ('base.html', 'base')
'''
Group(0) locates the whole match expression. 
group(1) means the first paranthesis pair locates matching expression 1
group(2) says the second next paranthesis pair locates the match expression 2, and so on.
'''
m = re.search('(?<=abc)(d(e))f', 'abcdef')
print(m.group(0))
print(m.group(1))
print(m.group(2))

# find inner pairs
innerre = re.compile("\('([^']+)', '([^']+)'\)")

results = innerre.findall(inner_str)
print(results)
for x,y in results:
    print("%s <-> %s" % (x,y))
```
